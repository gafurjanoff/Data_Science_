```python
data = dict()


data['Year'] = list(range(2011,2025))
data['Population'] = list(range(25,39))
data['Temp'] = [1.54,1.67,1.89,1.45,1.35,1.87,1.90,1.09,1.45,1.54,1.34,1.21,1.43,1.12]

data
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# @title Population vs Temp

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Population', y='Temp', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Year vs Population

from matplotlib import pyplot as plt
df.plot(kind='scatter', x='Year', y='Population', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
df.head(5)
```

```python
df.head()
```

```python
# @title Population

from matplotlib import pyplot as plt
df['Population'].plot(kind='hist', bins=20, title='Population')
plt.gca().spines[['top', 'right',]].set_visible(True)
```

```python
# @title Temp

from matplotlib import pyplot as plt
df['Temp'].plot(kind='hist', bins=20, title='Temp')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
