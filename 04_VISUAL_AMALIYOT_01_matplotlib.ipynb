## 8-VAZIFA. Toshkent shahri uchun, har bir yilda o'rtacha min va max haroratlarni ko'rsatuvchi grafik chizing. **Grafikni faylga saqlab oling**

```python
data = pd.read_csv('/content/Tashkent_Temprature.csv')
```

```python
data.rename(columns = {'Tm':'Tmin', 'TM':'Tmax'}, inplace = True)
```

```python
data.columns
```

```python
min = data['Tmin']
max = data['Tmax']
plt.figure(figsize = (10,5))
plt.plot(min, linewidth=2, linestyle="--", marker='h', markerfacecolor='g', label='Minimum Temprature')
plt.plot(max, linewidth=2, linestyle="-", marker='v', markerfacecolor='r', label='Maximum Temprature')
plt.legend(loc = 0)
plt.grid()
plt.title('Annual Temprature For Tashkent City')
plt.ylabel('Temprature')
plt.xlabel('Years')
plt.savefig('TempMinMax.png')
plt.show()

# plt.legend(['Minimum Temprature','Maximum Temprature'], loc=0, fontsize='x-large')
```
