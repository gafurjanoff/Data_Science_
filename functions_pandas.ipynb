```python
def change(e):
  return lambda e : np.max(e) + np.min(e)

result = df.applymap(change(df))
result
```

```python
def foo(x):
  return pd.Series([np.min(x) , np.max(x)] , index = ['Min' , 'Max'])

result = df.apply(foo)

print(result)
```
